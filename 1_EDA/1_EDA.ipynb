## 6.Detect and treat outliers for accurate analysis.
Finding the outliers from this dataset. Outliers make the data inconsistence which result in predection in accuracy for the models.So we need to handel them.

```python
sns.boxplot(df['Sales'])
```

### Insights:
- Sales is right-skewed, indicating some regions have very high sales
- Several extreme values are present beyond the upper whisker
- These outliers can significantly affect mean and variance
- Outliers may represent premium cabins and should not be blindly removed

```python
Q1 = df["Sales"].quantile(0.25)
Q3 = df["Sales"].quantile(0.75)

IQR = Q3- Q1

lower_bond = Q1 - 1.5* IQR
upper_bond = Q3 + 1.5* IQR
```

```python
df["Sales"] = np.where(df["Sales"] > upper_bond ,upper_bond , df["Sales"])
df["Sales"] = np.where(df["Sales"] < lower_bond , lower_bond , df["Sales"])
```

```python
sns.boxplot(df['Sales'])
```

## 11.Solving EDA questions on this dataset.

### 1.How has profit changed over the years?


```python
profit_year = df.groupby('order year')['profit'].sum().reset_index(name='Profit')
profit_year.columns = ['years','profit']
```

```python
plt.plot(profit_year['years'], profit_year['profit'], marker='o')
plt.title('Year-wise Profit Trend')
plt.xlabel('Year')
plt.ylabel('Total Profit')
```

#### Insights:
- Profit increases by the years.
- Year 2017 is the heighest profit with the profit more that 90000.
- From year 2015 to 2016 profit increased significantly.

### 2.Which category contributes the most to profit each year?

```python
tp = df.groupby(['order year', 'category'])['profit'].sum().reset_index()
```

```python
tp
```

```python
sns.barplot(x='order year', y='profit', hue='category', data=tp)

plt.title("Year-wise Profit by Category")
plt.xlabel("Year")
plt.ylabel("Profit")

```

#### Insights:
- Profit 'office supplies' and 'technology' category is increasing over the years.
- But profit of 'furniture' category is not increasing in a particular manner in year 2015 is decreases, in year 2016 it increases and again in 2017 it decreases.
- 'Technology' category produces heighest amount of profit every year than other categories, however in 2014 'office supplies' produces heighest profit than other categories that year.

### 3.Which region shows the highest growth in profit over time?

```python
sns.barplot(x='order year', y='profit', hue='region', data=r_pot)

plt.title("Year-wise Profit by Region")
plt.xlabel("Year")
plt.ylabel("Profit")
```

#### Insights:
- "East" and "West" region show continous growth over the years.
- "West" region shows heighest growth over the years than other regions.
- "Central" region started showing growth from 2015.

### 4. How does discount impact profit across categories?

- Rest of the sub-categories are profitable for every year.
