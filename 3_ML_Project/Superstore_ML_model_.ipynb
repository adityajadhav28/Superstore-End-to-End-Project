### Models Evaluation

```python
from sklearn.metrics import accuracy_score,confusion_matrix,precision_score,recall_score,f1_score
```

```python
def evaluate_model(y_pred,y_test,model_name):
    print("Model:",model_name)
    acc = accuracy_score(y_pred,y_test)
    prec = precision_score(y_pred,y_test)
    rec = recall_score(y_pred,y_test)
    f1 = f1_score(y_pred,y_test)
    print("Accuracy Score:",acc*100,"%")
    print("Precision Score:",prec*100,"%")
    print("Recall Score:",rec*100,"%")
    print("F1 Score:",f1*100,"%")
    print("\nConfusion Matrix:")
    print(confusion_matrix(y_pred,y_test))
    print("\n----------------------------------------------------------\n")
```

```python
evaluate_model(svm_pred,y_test,"SVM")
evaluate_model(lr_pred,y_test,"Logistic Regression")
evaluate_model(rf_pred,y_test,"Random Forest")
evaluate_model(ada_pred,y_test,"Ada Boost")
```
